```python
from os import listdir
from xml.etree import ElementTree
from numpy import zeros, asarray
from mrcnn.utils import Dataset
from mrcnn.config import Config
from mrcnn.model import MaskRCNN
from mrcnn import visualize
```

```python
# class that defines and loads the cell dataset
class CellDataset(Dataset):
    # load the dataset definitions
    def load_dataset(self, dataset_dir, is_train=True):
        # define one class
        self.add_class("dataset", 1, "cell")
        # image tags
        imgs = [2,6,7,15,17]
        # find all images
        # Traverse train data
        if is_train:
            for files in imgs:
                images_dir = dataset_dir + '/Training_Data_Set/image_' + str(files) +'/'
                annotations_dir = dataset_dir + '/Training_Data_Set/image_' + str(files) + '_XML/'   
                for filename in listdir(images_dir):
                    # extract image id
                    image_id = filename[:-4] 
                    img_path = images_dir + filename
                    ann_path = annotations_dir + image_id + '.xml'
                    # add to dataset
                    self.add_image('dataset', image_id=image_id, path=img_path, annotation=ann_path) 
        # Traverse test data
        if not is_train: 
            images_dir = dataset_dir + '/Test_Data_Set/image_20/'
            annotations_dir = dataset_dir + '/Test_Data_Set/image_20_XML/'
            for filename in listdir(images_dir):
                image_id = filename[:-4]
                img_path = images_dir + filename
                ann_path = annotations_dir + image_id + '.xml'
                # add to dataset
                self.add_image('dataset', image_id=image_id, path=img_path, annotation=ann_path)
    # Used to detect cells on a full size image
    def load_final(self, dataset_dir):
        self.add_class("dataset", 1, "cell")
        images_dir = dataset_dir + '/Original_Cell_Images/'
        annotations_dir = dataset_dir + '/Original_XML_Annotations/'
        for filename in listdir(images_dir):
            image_id = filename[:-4]
            img_path = images_dir + filename
            ann_path = annotations_dir + image_id + '.xml'
            # add to dataset
            self.add_image('dataset', image_id=image_id, path=img_path, annotation=ann_path)
    
    # extract bounding boxes from an annotation/xml file
    def extract_boxes(self, filename):
        # load and parse the file
        tree = ElementTree.parse(filename)
        # get the root of the document
        root = tree.getroot()
        # extract each bounding box
        boxes = list()
        for box in root.findall('.//bndbox'):
            #xmin = int(box.find('xmin').text)
            #ymin = int(box.find('ymin').text)
            #xmax = int(box.find('xmax').text)
            #ymax = int(box.find('ymax').text)
            xmin = int(round(float(box.find('xmin').text)))
            ymin = int(round(float(box.find('ymin').text)))
            xmax = int(round(float(box.find('xmax').text)))
            ymax = int(round(float(box.find('ymax').text)))
            coords = [xmin, ymin, xmax, ymax]
            boxes.append(coords)
        # extract image dimensions
        width = int(root.find('.//size/width').text)
        height = int(root.find('.//size/height').text)
        return boxes, width, height
 
    # load the masks for an image
    def load_mask(self, image_id):
        # get details of image
        info = self.image_info[image_id]
        # define box file location
        path = info['annotation']
        # load XML
        boxes, w, h = self.extract_boxes(path)
        # create one array for all masks, each on a different channel
        masks = zeros([h, w, len(boxes)], dtype='uint8')
        # create masks
        class_ids = list()
        for i in range(len(boxes)):
            box = boxes[i]
            row_s, row_e = box[1], box[3]
            col_s, col_e = box[0], box[2]
            masks[row_s:row_e, col_s:col_e, i] = 1
            class_ids.append(self.class_names.index('cell'))
        return masks, asarray(class_ids, dtype='int32')
 
    # load an image reference
    def image_reference(self, image_id):
        info = self.image_info[image_id]
        return info['path']
    
# define a configuration for the model
class CellConfig(Config):
    # define the name of the configuration
    NAME = "cell_cfg"
    # number of classes (background + cell)
    NUM_CLASSES = 1 + 1
    # number of training steps per epoch
    STEPS_PER_EPOCH = 80 
    # 0.002 is reasonable, 0.02 is not
    LEARNING_RATE = 0.002
    # Dimension specification allows faster training
    IMAGE_MIN_DIM = 256
    IMAGE_MAX_DIM = 256
    # 2 images per GPU works fine
    #IMAGES_PER_GPU = 1
    # Validation steps match the number of testing images
    VALIDATION_STEPS = 16
    # May leave this config
    #TRAIN_ROIS_PER_IMAGE = 200
```

```python
#Directory is in current directory
cell_dir = './'
# prepare train set
train_set = CellDataset()
train_set.load_dataset(cell_dir, is_train=True)
train_set.prepare()
print('Training images: %d' % len(train_set.image_ids))
# prepare test/val set
test_set = CellDataset()
test_set.load_dataset(cell_dir, is_train=False)
test_set.prepare()
print('Test images: \t%d' % len(test_set.image_ids))
# prepare config
config = CellConfig()
config.display()
```

```python
# define the model for training
model = MaskRCNN(mode='training', model_dir='./', config=config)
# load weights (mscoco) and exclude the output layers
model.load_weights('mask_rcnn_cell_cfg_0009.h5', by_name=True, exclude=["mrcnn_class_logits", "mrcnn_bbox_fc",  "mrcnn_bbox", "mrcnn_mask"])
#model.load_weights('mask_rcnn_coco.h5', by_name=True, exclude=["mrcnn_class_logits", "mrcnn_bbox_fc",  "mrcnn_bbox", "mrcnn_mask"])
```

```python
# train weights (output layers or 'heads')
model.train(train_set, test_set, learning_rate=config.LEARNING_RATE, epochs=10, layers='heads')
```

```python
from numpy import mean, expand_dims
from mrcnn.utils import compute_ap
from mrcnn.model import load_image_gt, mold_image
from matplotlib import pyplot
from matplotlib.patches import Rectangle
```

```python
# calculate the mAP for a model on a given dataset
def evaluate_model(dataset, model, cfg):
    APs = list()
    for image_id in dataset.image_ids:
        # load image, bounding boxes and masks for the image id
        image, image_meta, gt_class_id, gt_bbox, gt_mask = load_image_gt(dataset, cfg, image_id, use_mini_mask=False)
        # convert pixel values (e.g. center)
        scaled_image = mold_image(image, cfg)
        # convert image into one sample
        sample = expand_dims(scaled_image, 0)
        # make prediction
        yhat = model.detect(sample, verbose=0)
        # extract results for first sample
        r = yhat[0]
        # calculate statistics, including AP
        AP, _, _, _ = compute_ap(gt_bbox, gt_class_id, gt_mask, r["rois"], r["class_ids"], r["scores"], r['masks'])
        # append APs
        APs.append(AP)
    # calculate the mean AP across all images
    mAP = mean(APs)
    return mAP  
```

```python
# plot a number of photos with ground truth and predictions (5 images)
def plot_actual_vs_predicted(dataset, model, cfg, n_images=5):
    # load image and mask
    for i in range(n_images):
        # load the image and mask
        image = dataset.load_image(i)
        mask, _ = dataset.load_mask(i)
        # convert pixel values (e.g. center)
        scaled_image = mold_image(image, cfg)
        # convert image into one sample
        sample = expand_dims(scaled_image, 0)
        # make prediction
        yhat = model.detect(sample, verbose=0)[0]
        # define subplot
        pyplot.subplot(n_images, 2, i*2+1)
        # plot raw pixel data
        pyplot.imshow(image)
        pyplot.title('Actual')
        # plot masks using alpha values
        for j in range(mask.shape[2]):
            pyplot.imshow(mask[:, :, j], cmap='gray', alpha=0.3)
        # get the context for drawing boxes
        pyplot.subplot(n_images, 2, i*2+2)
        # plot raw pixel data
        pyplot.imshow(image)
        pyplot.title('Predicted')
        ax = pyplot.gca()
        # plot each box
        for box in yhat['rois']:
            # get coordinates
            y1, x1, y2, x2 = box
            # calculate width and height of the box
            width, height = x2 - x1, y2 - y1
            # create the shape
            rect = Rectangle((x1, y1), width, height, fill=False, color='red')
            # draw the box
            ax.add_patch(rect)
    # show the figure
    pyplot.show()
```

```python
from keras.preprocessing.image import img_to_array, array_to_img, load_img
import numpy as np
```

```python
# returns the predicted data from an input image
def gather_data(dataset, model, cfg, image_num):
    # load the image and mask
    image = dataset.load_image(image_num)
    mask, _ = dataset.load_mask(image_num)
    # convert pixel values (e.g. center)
    scaled_image = mold_image(image, cfg)
    # convert image into one sample
    sample = expand_dims(scaled_image, 0)
    # make prediction
    yhat = model.detect(sample, verbose=0)[0]
    return yhat
```

```python
# Plots boxes 
def plot_boxes(yhat):
    # counts the amount of boxes as it progresses
    box_count = 0
    pyplot.title('Prediction')
    pyplot.imshow(image)
    ax = pyplot.gca()
    # loop through rois in model detection
    for box in yhat['rois']:
        # get coordinates
        y1, x1, y2, x2 = box
        # calculate width and height of the box
        width, height = x2 - x1, y2 - y1
        # removes random bad guesses
        if ((width < 75) and (height < 75)):
            # create the shape
            rect = Rectangle((x1, y1), width, height, fill=False, color='red')
            # draw the box
            ax.add_patch(rect)   
            # count number of detections
            box_count += 1
    # show the figure
    pyplot.show()
    # display total detections
    print("Number of cells detected: " + str(box_count))
```

```python
# Final predictions visualized - Function not working correctly with multiple images - works with single
test_images = CellDataset()
# path of images
#images_path = cell_dir + 'Yuliya_Data'
# directory with a single image
images_path = cell_dir + 'Y_D' 
test_images.load_final(images_path)
test_images.prepare()
# choose single image to predict for
image_num = 0
# config needs to be for full size images
yhat = gather_data(test_images,model,cfg,image_num)
plot_boxes(yhat)
```

```python
# this function takes the broken down images and puts them back together as one
def stitch_up(images):
    # works through blocks of 4 on the x axis
    img_array = np.append(images[0],images[1],axis=1)
    for ii in range(1,3):
        img_array = np.append(img_array,images[ii+1],axis=1)

    img_array_2 = np.append(images[4],images[5],axis=1)
    for ii in range(5,7):
        img_array_2 = np.append(img_array_2,images[ii+1],axis=1)
    img_array_app = np.append(img_array,img_array_2,axis=0)

    img_array_3 = np.append(images[8],images[9],axis=1)
    for ii in range(9,11):
        img_array_3 = np.append(img_array_3,images[ii+1],axis=1)
    img_array_app = np.append(img_array_app,img_array_3,axis=0)

    img_array_4 = np.append(images[12],images[13],axis=1)
    for ii in range(13,15):
        img_array_4 = np.append(img_array_4,images[ii+1],axis=1)
    img_array_app = np.append(img_array_app,img_array_4,axis=0)    
    img_result = array_to_img(img_array_app)
    # returns full size image
    return img_result
```

```python
# Plots boxes - iterates over the broken down image (into 16 images)
def plot_boxes_old(yhats):
    box_count = 0
    counter = 0
    image = stitch_up(images)
    pyplot.title('Stitched Image')
    pyplot.imshow(image)
    ax = pyplot.gca()
    for jj in range(4):
        for ii in range(4):
            # loop throguh rois in model detection
            for box in yhats[counter]['rois']:
                # get coordinates
                y1, x1, y2, x2 = box
                # calculate width and height of the box
                width, height = x2 - x1, y2 - y1
                # create the shape
                rect = Rectangle((x1+(56*ii), y1+(56*jj)), width, height, fill=False, color='red')
                # draw the box
                ax.add_patch(rect)   
                # count number of detections
                box_count += 1
            counter += 1
    # show the figure
    pyplot.show()
    # display total detections
    print("Number of cells detected: " + str(box_count))
```

```python
# returns the predicted data from input images
def gather_data_old(dataset, model, cfg, n_images=16):
    # same as other gatherData function but works with small images 
    yhats = list()
    images = list()
    for i in range(n_images):
        # load the image and mask
        image = dataset.load_image(i)
        mask, _ = dataset.load_mask(i)
        # convert pixel values (e.g. center)
        scaled_image = mold_image(image, cfg)
        # convert image into one sample
        sample = expand_dims(scaled_image, 0)
        # make prediction
        yhat = model.detect(sample, verbose=0)[0]
        yhats.append(yhat)
        images.append(image)
    return yhats,images
```
